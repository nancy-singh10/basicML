# Importing libb


```python
import numpy as np
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

# data colelection and process

```python
#loading dataset in pandas datafrom  no header in file
sonar=pd.read_csv("/content/Copy of sonar data.csv",header=None)
```

```python
sonar.head()
```

```python
#number of rows and col
sonar.shape
```

```python
#statics of data
sonar.describe()
```

```python
sonar[60].value_counts()
#find how many rocks and how any mines are their by  value count
```

#M--mine
#R--rock

```python
#group data o  mine and rock  --giving mean
sonar.groupby(60).mean()
```

```python
#seaprate data and Labels --r and m    axis 1 specidy drop col with label 60
#axis 0  row
#axis 1 col
X=sonar.drop(columns=60,axis=1)
y=sonar[60]
```

```python
X
```

```python
y
```

# Split in training and test data
**bold text**

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.1,stratify=y, random_state=1)
#test size 1 percent only to test
# test_size=0.1: This means 10% of the data will be used for the test set.
# stratify=y: This ensures that the proportion of classes in the target variable y is maintained in both the training and test sets.
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

# Model Training -->Logistic Regression

```python
model=LogisticRegression()
```

```python
#tain the logistic regression model with train data
model.fit(X_train,y_train)
```

```python
#accuracy on trainin data     70>>>> good
X_train_pred=model.predict(X_train)
train_Acc=accuracy_score(X_train_pred,y_train)

print("accuracy",train_Acc)
```

```python
#accuracy on test data     70>>>> good
X_test_pred=model.predict(X_test)
test_Acc=accuracy_score(X_test_pred,y_test)

print("accuracy",test_Acc)
```
